# q1.Scrape the details of most viewed videos on YouTube from Wikipedia:

```python
#open instance of browser
driver=webdriver.Chrome(r'chromedriver.exe')
url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
#navigate to required url
driver.get(url)
rank=[]
title=[]
artist=[]
view=[]
date=[]
time.sleep(3)
#scrape data from each match tab
nums=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
titles=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
artists=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
views=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
dates=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
#itenrate over teh elements to get the required data
for i in [nums,titles,artists,views,dates]:
    for j in i:
        if i == nums:
            rank.append(j.text)
        if i == titles:
            title.append(j.text)
        if i== artists:
            artist.append(j.text)
        if i == views:
            view.append(j.text)
        if i == dates:
            date.append(j.text)
#create data frame of the scraped data
youtube=pd.DataFrame({})
youtube['rank']=rank
youtube['title']=title
youtube['artist']=artist
youtube['views in billons']=view
youtube['date']=date
youtube
```

# q2.Scrape the details team India’s international fixtures from bcci.tv

```python
#open instance of teh browser
driver=webdriver.Chrome(r'chromedriver.exe')
url='https://www.bcci.tv/'

#get the required url
driver.get(url)

#procedure to click on the fixtures and navigate to internantional fixtures page
driver.find_element_by_xpath('//nav[@class="navigation__menu js-navigation-menu"]/ul/li[1]').click()
time.sleep(2)
driver.find_element_by_xpath('//nav[@class="navigation__menu js-navigation-menu"]/ul/li[1]/div/div/ul/li[1]').click()
time.sleep(3)
title=[]
tournament=[]
place=[]
time1=[]
date=[]

#scrape the required tabs
titles=driver.find_elements_by_xpath('//strong[@class="fixture__name fixture__name--with-margin"]')
tournaments=driver.find_elements_by_xpath('//div[@class="fixture__format-strip"]')
places=driver.find_elements_by_xpath('//p[@class="fixture__additional-info"]/span')
dates=driver.find_elements_by_xpath('//div[@class="fixture__datetime desktop-only"]/div/span')
months=driver.find_elements_by_xpath('//div[@class="fixture__datetime desktop-only"]/div/div/span[1]')
times=driver.find_elements_by_xpath('//div[@class="fixture__full-date"]/div/span[2]')

#iterate over each list to get the text
for i in tournaments:
    tournament.append(i.text)
for i in titles:
    title.append(i.text)
for i in places:
    place.append(i.text)
for i in times:
    time1.append(i.text)
for i in range(len(dates)):
    date.append(dates[i].text+months[i].text)

#create dataframe
icc=pd.DataFrame({})
icc['title']=title
icc['tournament']=tournament
icc['place']=place
icc['time']=time1
icc['date']=date
icc
```

# q4.Scrape the details of State-wise GDP of India from statisticstime.com.

```python
#create new instance of teh browser
driver=webdriver.Chrome(r'chromedriver.exe')
url='http://statisticstimes.com/'

#navigate to required url
driver.get(url)
driver.maximize_window()

#click on teh dropdown menu and navigate to the statewise gdp page
driver.find_element_by_xpath('//div[@class="navbar"]/div[2]').click()
time.sleep(1)
driver.find_element_by_xpath('//div[@class="navbar"]/div[2]/div/a[3]').click()
time.sleep(2)
driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[2]/ul/li[1]/a').click()
time.sleep(2)
rank=[]
state=[]
gdsp_1920=[]
gdsp_1819=[]
share=[]
gdp=[]

#scrape the required table
ranks=driver.find_elements_by_xpath('//div[@id="main"]/div[5]/div/div/table/tbody/tr/td[1]')
states=driver.find_elements_by_xpath('//div[@id="main"]/div[5]/div/div/table/tbody/tr/td[2]')
gdsp_1920s=driver.find_elements_by_xpath('//div[@id="main"]/div[5]/div/div/table/tbody/tr/td[3]')
gdsp_1819s=driver.find_elements_by_xpath('//div[@id="main"]/div[5]/div/div/table/tbody/tr/td[4]')
shares=driver.find_elements_by_xpath('//div[@id="main"]/div[5]/div/div/table/tbody/tr/td[5]')
gdps=driver.find_elements_by_xpath('//div[@id="main"]/div[5]/div/div/table/tbody/tr/td[6]')
tags=[ranks,states,gdsp_1920s,gdsp_1819s,shares,gdps]

#iterate on each tags list to get data 
for i in tags:
    for j in i:
        if i==tags[0]:
            rank.append(j.text)
        if i==tags[1]:
            state.append(j.text)
        if i==tags[2]:
            gdsp_1920.append(j.text)
        if i==tags[3]:
            gdsp_1819.append(j.text)
        if i==tags[4]:
            share.append(j.text)
        if i==tags[5]:
            gdp.append(j.text)

#create dataframe
economy=pd.DataFrame({})
economy['rank']=rank
economy['state']=state
economy['gdsp_1920']=gdsp_1920
economy['gdsp_1819']=gdsp_1819
economy['share']=share
economy['gdp']=gdp
economy
```

# q5.Scrape the details of trending repositories on Github.com

```python
driver=webdriver.Chrome(r'chromedriver.exe')
url='https://github.com/'
driver.get(url)
driver.maximize_window()
time.sleep(3)
driver.find_element_by_xpath('//ul[@class="d-lg-flex list-style-none"]/li[4]').click()
driver.find_element_by_xpath('//ul[@class="d-lg-flex list-style-none"]/li[4]').click()
time.sleep(2)
rep_name=[]
description=[]
contributor=[]
languages=[]
urls=[]
driver.find_element_by_xpath('//ul[@class="d-lg-flex list-style-none"]/li[4]/details/div/ul[2]/li[3]').click()
time.sleep(3)
rep_names=driver.find_elements_by_xpath('//article[@class="Box-row"]/h1/a')
descriptions=driver.find_elements_by_xpath('//article[@class="Box-row"]/p')
for i in [rep_names,descriptions]:
    for j in i:
        if i ==rep_names:
            rep_name.append(j.text)
            urls.append(j.get_attribute('href'))
        if i==descriptions:
            description.append(j.text)
    #rep_name.append(i.text)
#for i in rep_names:
    #urls.append(i.get_attribute('href'))
for url in urls:
    driver.get(url)
    time.sleep(4)
    div_list=driver.find_elements_by_xpath('//div[@class="BorderGrid BorderGrid--spacious"]/div')
    try:
        contributor.append(((div_list[-2].text).split())[1])
    except:
        contributor.append('-')
    try:
        languages.append(((div_list[-1].text).split())[1::2])
    except:
        languages.append('-')
github=pd.DataFrame({})
github['rep_name']=rep_name
github['description']=description
github['contributor']=contributor
github['languages']=languages
github.head()
```

# q6.Scrape the details of top 100 songs on billiboard.com.

```python
#open new instance of teh browser
driver=webdriver.Chrome(r'chromedriver.exe')
url='https://www.billboard.com/'

#navigate to required url
driver.get(url)
driver.maximize_window()
time.sleep(3)

#click on top100 tab
driver.find_element_by_xpath('//*[@id="root"]/div[2]/div[2]/nav/ul/li[3]/a').click()
name=[]
artist=[]
last_week_rank=[]
peak_rank=[]
no_of_week=[]

#scrape teh required data
song_names=driver.find_elements_by_xpath('//ol[@class="chart-list__elements"]/li/button/span[2]/span[1]')
artists=driver.find_elements_by_xpath('//ol[@class="chart-list__elements"]/li/button/span[2]/span[2]')
last_weeks=driver.find_elements_by_xpath('//ol[@class="chart-list__elements"]/li/button/div/div[1]')
peaks=driver.find_elements_by_xpath('//ol[@class="chart-list__elements"]/li/button/div/div[2]')
no_of_weeks=driver.find_elements_by_xpath('//ol[@class="chart-list__elements"]/li/button/div/div[3]')
for i in [song_names,artists,last_weeks,peaks,no_of_weeks]:
    for j in i:
        if i==song_names:
            name.append(j.text)
        if i==artists:
            artist.append(j.text)
        if i==last_weeks:
            last_week_rank.append(j.text)
        if i==peaks:
            peak_rank.append(j.text)
        if i==no_of_weeks:
            no_of_week.append(j.text)

#create a dataframe of teh acquired data            
billboard=pd.DataFrame({})
billboard['name']=name
billboard['artist']=artist
billboard['last_week_rank']=last_week_rank
billboard['peak_rank']=peak_rank
billboard['no_of_week']=no_of_week
billboard.head()            
```

# q9.Scrape the details most watched tv series of all time from imdb.com.

```python
#create new instance of the browser
driver=webdriver.Chrome(r'chromedriver.exe')

#navigate to required url
url='https://www.imdb.com/list/ls095964455/'
driver.get(url)
title=[]
span=[]
genre=[]
run_time=[]
rating=[]
vote=[]

#scrape the details from the table
titles=driver.find_elements_by_xpath('//div[@class="lister-item-content"]/h3/a')
spans=driver.find_elements_by_xpath('//div[@class="lister-item-content"]/h3/span[2]')
genres=driver.find_elements_by_xpath('//div[@class="lister-item-content"]/p/span[5]')
run_ts=driver.find_elements_by_xpath('//div[@class="lister-item-content"]/p/span[3]')
ratings=driver.find_elements_by_xpath('//div[@class="lister-item-content"]/div/div/span[2]')
votes=driver.find_elements_by_xpath('//div[@class="lister-item-content"]/p[4]/span[2]')

#iterate over each tags list to get text 
for i in [titles,spans,genres,run_ts,ratings,votes]:
    for j in i:
        if i==titles:
            title.append(j.text)
        if i==spans:
            span.append(j.text)
        if i==genres:
            genre.append(j.text)
        if i==run_ts:
            run_time.append(j.text)
        if i==ratings:
            rating.append(j.text)
        if i==votes:
            vote.append(j.text)
#create dataframe of teh acquired data 
imdb_shows=pd.DataFrame({})
imdb_shows['title']=title
imdb_shows['span']=span
imdb_shows['genre']=genre
imdb_shows['run_time']=run_time
imdb_shows['rating']=rating
imdb_shows['vote']=vote
imdb_shows.head()
```

# q10.Details of Datasets from UCI machine learning repositories

```python
#create new instance of the browser
driver=webdriver.Chrome(r'chromedriver.exe')

#navigating to required url
url='https://archive.ics.uci.edu/ml/index.php'
driver.get(url)
time.sleep(2)

#going to all datasets page
driver.find_element_by_xpath('/html/body/table[1]/tbody/tr/td[2]/span[2]').click()
time.sleep(3)

#scraping the table taht contains required data
names=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td/table/tbody/tr/td[2]')
data_types=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p')
tasks=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p')
attr_types=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p')
instances=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p')
attributes=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]/p')
years=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p')

name=[]
data_type=[]
task=[]
attr_type=[]
instance=[]
attribute=[]
year=[]

#iterating over teh tags list to get the required data
for i in names:
    name.append(i.text)
for i in [data_types,tasks,attr_types,instances,attributes,years]:
    for j in i[1:]:
        if i==data_types:
            data_type.append(j.text)
        if i==tasks:
            task.append(j.text)
        if i==attr_types:
            attr_type.append(j.text)
        if i==instances:
            instance.append(j.text)
        if i==attributes:
            attribute.append(j.text)
        if i==years:
            year.append(j.text)

#creating dataframe of teh acquired data
datasets=pd.DataFrame({})
datasets['names']=name
datasets['data_types']=data_type
datasets['tasks']=task
datasets['attr_types']=attr_type
datasets['instances']=instance
datasets['attributes']=attribute
datasets['years']=year
datasets.head()
```

# q7.Scrape the details of Data science recruiters from naukri.com.

```python
#creating new instance of teh browser
driver=webdriver.Chrome(r'chromedriver.exe')

#navigating to required url
url='https://www.naukri.com/hr-recruiters-consultants'
driver.get(url)

#entering keyword in search box
driver.find_element_by_xpath('//input[@class="sugInp"]').send_keys('Data Science')

#clicking search button
driver.find_element_by_xpath('//button[@class="fl qsbSrch blueBtn"]').click()
time.sleep(3)
name=[]
position=[]
company=[]
location=[]
skill=[]
names=driver.find_elements_by_xpath('//div[@class="recInfo"]/div[1]/p/a[1]')
urls=[]
for i in names:
    name.append(i.text)
    urls.append(i.get_attribute('href'))
#navigating to each page and scraping the required data
for url in urls:
    driver.get(url)
    position.append(driver.find_element_by_xpath('//div[@class="rFrame fl infoWrapper"]/div[3]').text)
    company.append(driver.find_element_by_xpath('//div[@class="rFrame fl infoWrapper"]/div[4]').text)
    try:
        location.append(driver.find_element_by_xpath('//div[@class="rFrame fl infoWrapper"]/div[3]').text)
    except:
        location.append('-')
    skill.append(driver.find_element_by_xpath('//div[@class="fadeOutParent readMoreHt100 "]/div[2]/div/p').text)
    
#creating the dataframe of teh scraped data            
recruiters=pd.DataFrame({})
recruiters['name']=name
recruiters['position']=position
recruiters['company']=company
recruiters['location']=location
recruiters['skill']=skill
recruiters.head()
```
