```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
```

# q1: data analyst jobs in bangalore

```python
def naukri_jobs(url):
    driver = webdriver.Chrome(r'chromedriver.exe')
    driver.get(url)
    time.sleep(3)
    title=[]
    company=[]
    experience=[]
    salary=[]
    driver.find_element_by_xpath('//input[@class="sugInp"]').send_keys('Data Analyst')
    driver.find_element_by_xpath('//input[@id="qsb-location-sugg"]').send_keys('Bangalore')
    driver.find_element_by_xpath('//button[@class="btn"]').click()
    time.sleep(5)
    titles=driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
    for i in titles:
        title.append(i.text)
    companies=driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]')
    for i in companies:
        company.append(i.text)
    exp=driver.find_elements_by_xpath('//span[@class="ellipsis fleft fs12 lh16"]')
    for i in exp:
        experience.append(i.text)
    salaries=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi salary"]')
    for i in salaries:
        salary.append(i.text)
    jobs_da=pd.DataFrame({})
    jobs_da['Title']=title[0:10]
    jobs_da['Company']=company[0:10]
    jobs_da['Experience']=experience[0:10]
    jobs_da['Salary']=salary[0:10]
    return jobs_da
    jobs_da.to_csv('Data Analyst Jobs')
```

```python
naukri_jobs('https://www.naukri.com')
```

# q2:data scientist jobs with full description

```python
def naukri_jobs_ds(url):
    driver = webdriver.Chrome(r'chromedriver.exe')
    driver.get(url)
    driver.find_element_by_xpath('//input[@class="sugInp"]').send_keys('Data Scientist')
    driver.find_element_by_xpath('//input[@id="qsb-location-sugg"]').send_keys('Bangalore')
    title=[]
    company=[]
    location=[]
    job_desc=[]
    driver.find_element_by_xpath('//button[@class="btn"]').click()
    time.sleep(5)
    urls=[]
    titles=driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
    for i in titles:
        url=i.get_attribute('href')
        urls.append(url)
    for url in urls:
        driver.get(url)
        time.sleep(3)
        try:
            k=driver.find_element_by_xpath('//h1[@class="jd-header-title"]')
            title.append(k.text)
        except:
            pass
        try:
            k=driver.find_element_by_xpath('//span[@class="location "]')
            location.append(k.text)
        except:
            pass
        try:
            k=driver.find_element_by_xpath('//div[@class="jd-header-comp-name"]')
            company.append(k.text)
        except:
            pass
        try:
            k=driver.find_element_by_xpath('//div[@class="dang-inner-html"]')
            job_desc.append(k.text.replace('\n',''))
        except:
            pass
    jobs_one=pd.DataFrame({})
    jobs_one['Title']=title
    jobs_one['Company']=company
    jobs_one['Location']=location
    jobs_one['JOB_Desc']=job_desc
    return jobs_one
    jobs_one.to_csv('Jobs_full_desc')
```

```python
naukri_jobs_ds('https://www.naukri.com')
```

# q7: iphone reviews

```python
def iphone_review(url):
    driver = webdriver.Chrome(r'chromedriver.exe')
    driver.get(url)
    time.sleep(3)
    star_rating=[]
    short_review=[]
    full_review=[]
    j=0
    while j<10:
    
        stars=driver.find_elements_by_xpath('//div[@class="_3LWZlK _1BLPMq"]')
        for i in stars:
            star_rating.append(i.text)
        con_reviews=driver.find_elements_by_xpath('//p[@class="_2-N8zT"]')
        for i in con_reviews:
            short_review.append(i.text)
        full_reviews=driver.find_elements_by_xpath('//div[@class="t-ZTKy"]')
        for i in full_reviews:
            full_review.append(i.text)
        driver.find_element_by_xpath('//a[@class="_1LKTO3"]').click()
        j+=1
        time.sleep(3)
    reviews_iphone=pd.DataFrame({})
    reviews_iphone['Stars']=star_rating
    reviews_iphone['Short Review']=short_review
    reviews_iphone['Detailed Reviews']=full_review
    return reviews_iphone
    reviews_iphone.to_csv('Iphone_reviews')
```

# q10: laptop info

```python
def laptop_info(url):
    driver = webdriver.Chrome(r'chromedriver.exe')
    driver.get(url)
    Title=[]
    Price=[]
    Rating=[]
    driver.find_element_by_xpath('//input[@id="twotabsearchtextbox"]').send_keys('Laptop')
    driver.find_element_by_xpath('//input[@type="submit"]').click()
    driver.find_element_by_xpath('//*[@id="p_n_feature_thirteen_browse-bin/12598163031"]/span/a').click()
    time.sleep(5)
    driver.find_element_by_xpath('//*[@id="p_n_feature_thirteen_browse-bin/16757432031"]/span/a').click()
    titles=driver.find_elements_by_xpath('//span[@class="a-size-medium a-color-base a-text-normal"]')
    for title in titles:
        Title.append(title.text)
    prices=driver.find_elements_by_xpath('//span[@class="a-price-whole"]')
    for price in prices:
        Price.append(price.text)

    ratings=driver.find_elements_by_xpath('//div[@class="a-row a-size-small"]/span[1]')   
    for rating in ratings:
        Rating.append(rating.get_attribute('aria-label'))
    laptops=pd.DataFrame({})
    laptops['Title']=Title[:11]
    laptops['Price']=Price[:11]
    laptops['Rating']=Rating[:11]
    return laptops
    laptops.to_csv('Laptops_info')
```

```python
laptop_info('https://www.amazon.in/')
```

# q6:sunglasses info

```python
def sunglasses(url):
    driver = webdriver.Chrome(r'chromedriver.exe')
    driver.get(url)
    driver.find_element_by_xpath('//button[@class="_2KpZ6l _2doB4z"]').click()
    time.sleep(3)
    driver.find_element_by_xpath('//input[@title="Search for products, brands and more"]').send_keys('sunglasses')
    driver.find_element_by_xpath('//button[@type="submit"]').click()
    time.sleep(3)
    brand_name=[]
    prod_desc=[]
    prod_price=[]
    prod_disc=[]
    j=0
    while j<3:
        time.sleep(3)
        brands=driver.find_elements_by_xpath('//div[@class="_2WkVRV"]')
        for brand in brands:
            brand_name.append(brand.text)
        descriptions=driver.find_elements_by_xpath('//a[contains(@class,"IRpwTa")]')
        for description in descriptions:
            prod_desc.append(description.text)
        prices=driver.find_elements_by_xpath('//div[@class="_30jeq3"]')
        for price in prices:
            prod_price.append(price.text)
        discounts=driver.find_elements_by_xpath('//div[@class="_3Ay6Sb"]')
        for discount in discounts:
            prod_disc.append(discount.text)
        if j==0:
            driver.find_element_by_xpath('//a[@class="_1LKTO3"]').click()
        elif j==1:
            driver.find_element_by_xpath('//nav[@class="yFHi8N"]/a[12]').click()
        else:
            pass
        j+=1
    sunglasses=pd.DataFrame({})
    sunglasses['Brand']=brand_name[:100]
    sunglasses['Prod_Desc']=prod_desc[:100]
    sunglasses['Prod_Price']=prod_price[:100]
    sunglasses['Prod_Disc']=prod_disc[:100]
    return sunglasses
    sunglasses.to_csv('Sunglasses.csv')
```

# q9:myntra shoes

```python
 def myntra_shoes(url):
    driver = webdriver.Chrome(r'chromedriver.exe')
    driver.get(url)
    time.sleep(5)
    brand_name=[]
    description=[]
    prod_price=[]
    driver.find_element_by_xpath('//ul[@class="price-list"]/li[2]').click()
    time.sleep(3)
    driver.find_element_by_xpath('//span[@data-colorhex="black"]').click()
    time.sleep(3)
    
    j=0
    while j<2:
        time.sleep(5)
        brands=driver.find_elements_by_xpath('//h3[@class="product-brand"]')
        for brand in brands:
            brand_name.append(brand.text)
        descs=driver.find_elements_by_xpath('//h4[@class="product-product"]')
        for desc in descs:
            description.append(desc.text)
        prices=driver.find_elements_by_xpath('//div[@class="product-price"]')
        for price in prices:
                prod_price.append(price.text)
        driver.find_element_by_xpath('//a[@rel="next"]').click()    
        
        j+=1
    myntra=pd.DataFrame({})
    myntra['Brand']=brand_name
    myntra['Prod_Desc']=description
    myntra['Prod_Price']=prod_price
    return myntra
    myntra.to_csv('Myntra_shoes')
```

# q4 : glass door jobs

```python
def glass_jobs(url):
    job_title=[]
    company_name=[]
    rating=[]
    salary=[]
    driver = webdriver.Chrome(r'chromedriver.exe')
    driver.get(url)
    driver.find_element_by_xpath('//div[@class="locked-home-sign-in"]').click()
    driver.find_element_by_xpath('//input[@id="userEmail"]').send_keys('abcxyzdotdotdot@gmail.com')
    driver.find_element_by_xpath('//input[@id="userPassword"]').send_keys('123ABC456xyz')
    driver.find_element_by_xpath('//button[@class="gd-ui-button minWidthBtn css-8i7bc2"]').click()
    driver.maximize_window()
    time.sleep(3)
    driver.find_element_by_xpath('//*[@id="sc.keyword"]').send_keys('Data Scientist')
    driver.find_element_by_xpath('//*[@id="sc.location"]').send_keys(Keys.CONTROL+'a')
    driver.find_element_by_xpath('//*[@id="sc.location"]').send_keys(Keys.DELETE)
    driver.find_element_by_xpath('//*[@id="sc.location"]').send_keys('Noida')
    time.sleep(3)
    driver.find_element_by_xpath('//*[@id="scBar"]/div/button').click()
    time.sleep(4)
    jobs=driver.find_elements_by_xpath('//article[@id="MainCol"]/div/ul/li')
    for job in jobs:
        try:
            driver.find_element_by_xpath('//*[@id="JAModal"]/div/div[2]/span').click()
        except:
            pass
        job.click()
        time.sleep(3)
        job_title.append((driver.find_element_by_xpath('//div[@class="css-1vg6q84 e1tk4kwz4"]')).text)
        company_name.append((driver.find_element_by_xpath('//div[@class="css-87uc0g e1tk4kwz1"]')).text)
        try:
            rating.append((driver.find_element_by_xpath('//span[@data-test="detailRating"]')).text)
        except:
            rating.append('Not Available')
        try:
            salary.append((driver.find_element_by_xpath('//span[@data-test="detailSalary"]')).text)
        except:
            pass
    jobs_glass=pd.DataFrame({})
    jobs_glass['Title']=job_title[0:10]
    jobs_glass['Company']=company_name[0:10]
    jobs_glass['Salary']=salary[0:10]
    jobs_glass['Rating']=rating[0:10]
    return jobs_glass
    jobs_glass.to_csv('glass_jobs')
```

# q5: glassdoor salary

```python
def salaries(url):
    min_sal=[]
    max_sal=[]
    avg_sal=[]
    company_name=[]
    rating=[]

    driver = webdriver.Chrome(r'chromedriver.exe')
    driver.get(url)
    driver.find_element_by_xpath('//*[@id="KeywordSearch"]').send_keys('Data Scientist')
    driver.find_element_by_xpath('//*[@id="LocationSearch"]').send_keys(Keys.CONTROL+'a')
    driver.find_element_by_xpath('//*[@id="LocationSearch"]').send_keys(Keys.DELETE)
    driver.find_element_by_xpath('//*[@id="LocationSearch"]').send_keys('Noida')
    time.sleep(3)
    driver.find_element_by_xpath('//button[@class="gd-btn-mkt"]').click()
    time.sleep(5)
    companies=driver.find_elements_by_xpath('//div[@data-test="job-info"]/p[2]')
    avg_salary=driver.find_elements_by_xpath('//div[@class="col-2 d-none d-md-flex flex-row justify-content-end"]/strong')
    min_salary=driver.find_elements_by_xpath('//div[@class="common__RangeBarStyle__values common__flex__justifySpaceBetween common__flex__container "]/span[1]')
    max_salary=driver.find_elements_by_xpath('//div[@class="common__RangeBarStyle__values common__flex__justifySpaceBetween common__flex__container "]/span[2]')
    for i in companies:
        company_name.append(i.text)
    for i in avg_salary:
        avg_sal.append(i.text)
    for i in min_salary:
        min_sal.append(i.text)
    for i in max_salary:
        max_sal.append(i.text)
    salaries=pd.DataFrame({})
    salaries['Company_Name']=company_name[:10]
    salaries['Avg_Sal']=avg_sal[:10]
    salaries['Min_Sal']=min_sal[:10]
    salaries['Max_Sal']=max_sal[:10]
    return salaries
    salaries.to_csv('Glass_Sal')
```

```python
salaries('https://www.glassdoor.co.in/Salaries/index.htm')
```
